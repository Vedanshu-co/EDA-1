```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"amazon.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Clean price and discount columns by removing '₹', ',', and '%'
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
df.head()
```

```python
df.drop(columns=(['product_id','product_link','img_link']))
```

```python
df.isnull().sum()
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '').fillna('0').astype(int)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce') 
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python

df['main_category'] = df['category'].astype(str).str.split('|').str[0]
df['sub_Category'] =df['category'].astype(str).str.split('|').str[1]
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].dtypes
```

```python
categories = df.groupby('main_category').agg(total_products=('product_id', 'count'),avg_rating=('rating', 'mean'),avg_discount=('discount_percentage', 'mean'),avg_discounted_price=('discounted_price', 'mean')).reset_index()

categories.sort_values(by='total_products', ascending=False)
```

```python
df.groupby('main_category')['discount_amount'].sum().sort_values(ascending = False).head()
```

```python
df.groupby('main_category')['discount_amount'].sum().sort_values(ascending = False).tail()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, y='main_category', order=df['main_category'].value_counts().index)
plt.title('Product Distribution by Main Category')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
df.groupby('sub_Category')['discount_amount'].sum().sort_values(ascending = False).head()
```

```python
df.groupby('sub_Category')['discount_amount'].sum().sort_values(ascending = False).tail()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=df, y='sub_Category', order=df['sub_Category'].value_counts().index)
plt.title('Product Distribution by Main Category')
plt.xlabel('Number of Products')
plt.ylabel('Category')
plt.show()
```

```python
sns.scatterplot( data=df, x='discount_percentage', y='rating', alpha=0.6, color='teal')
plt.title('Discount Percentage vs. Customer Rating')
plt.show()
```

```python
top_reviewed = df.sort_values(by='rating_count', ascending=False).head(10)
plt.figure(figsize=(15, 6))
sns.barplot(data=top_reviewed, x='rating_count', y='product_name', hue='product_name',palette='magma',legend=False)
plt.title('Top 10 Most Reviewed Products on Amazon')
plt.xlabel('Total Rating Count')
plt.ylabel('Product Name')
plt.show()
```
